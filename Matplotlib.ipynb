#Matplotlib

```
import matplotlib.pyplot as plt
import math
import numpy as np
```

```
plt.plot([1,2,3,4])#creates a 2D object
plt.show()#shows the plot
```

```
plt.plot([1,2,3,4],[1,4,9,16])#first argument -> x-axis, second -> y-axis
plt.show()
```

##Adding properties

```
plt.axis([0,5,0,20])#Set the extremes of the plot
plt.title('My first plot')
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.show()
```

```
t = np.arange(0,2.5,0.1)
y1 = np.sin(math.pi*t)
y2 = np.sin(math.pi*t + math.pi/2)
y3 = np.sin(math.pi*t - math.pi/2)
plt.plot(t,y1,'b^',t,y2,'g*',t,y3,'ys')
plt.show()
```

```
plt.plot(t,y1,'b--',t,y2,'g.',t,y3,'y-.')
plt.show()
```

```
plt.plot([1,2,4,2,1,0,1,2,1,4],linewidth = 2.0)
plt.show()
```

##Plotting two plots simultaneously

```
t = np.arange(0,5,0.1)
y1 = np.sin(2*np.pi*t)
y2 = np.cos(2*np.pi*t)
plt.subplot(211)
plt.plot(t,y1,'r--')
plt.subplot(212)
plt.plot(t,y2,'b.-')
plt.show()
```

```
t = np.arange(0,5,0.1)
y1 = np.sin(2*np.pi*t)
y2 = np.cos(2*np.pi*t)
plt.subplot(121)
plt.plot(t,y1,'r--')
plt.subplot(122)
plt.plot(t,y2,'b.-')
plt.show()
```

##Adding text

```
x = np.arange(0,5,1)
y = [0,1,4,9,16]
plt.axis([0,4,0,16])
plt.title('my second plot')
plt.xlabel('number')
plt.ylabel('square')
plt.plot(x,y,'ro')
plt.show()
```

```
x = np.arange(0,5,1)
y = [0,1,4,9,16]
plt.axis([0,4,0,16])
plt.title('my second plot',fontsize = 20,color='gray')
plt.xlabel('number',color='green')
plt.ylabel('square',color='blue')
plt.plot(x,y,'ro')
plt.show()
```
